
# Linear Regression Implementation

This notebook demonstrates **Simple Linear Regression** using the **House Size vs House Price** example from the notes.

## Objective
Predict house prices based on house size using a best-fit line.

### Dataset

| House Size (sq. ft.) | Price ($) |
|---|---:|
| 800 | 120,000 |
| 1000 | 150,000 |
| 1200 | 180,000 |
| 1500 | 225,000 |

The goal is to learn the relationship:

\[ \hat{y} = wx + b \]

where:

- **x** = House size
- **w** = Coefficient (slope)
- **b** = Intercept (bias)
- **ŷ** = Predicted price


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    'House_Size': [800, 1000, 1200, 1500],
    'Price': [120000, 150000, 180000, 225000]
}

df = pd.DataFrame(data)
df
```


## Step 1: Visualize the Data

Before training a model, always visualize the relationship between the input feature and target variable.

If the points roughly follow a straight line, linear regression is a good choice.


```python

plt.figure(figsize=(8,5))
plt.scatter(df['House_Size'], df['Price'], s=100)

plt.title('House Size vs Price')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.grid(True)

plt.show()
```


## Step 2: Prepare Features and Target

- **X** contains the input feature (house size).
- **y** contains the target variable (price).


```python

X = df[['House_Size']]
y = df['Price']

print("X shape:", X.shape)
print("y shape:", y.shape)
```


## Step 3: Train the Linear Regression Model

We use the `LinearRegression` class from Scikit-learn.

During training, the algorithm learns:

- Slope (coefficient)
- Intercept

These values define the best-fitting line.


```python

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("Coefficient (w):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```


## Step 4: Model Equation

The trained model follows:

\[ \hat{y} = wx + b \]

Substitute the learned values:

\[ \hat{y} = ({:.2f})x + ({:.2f}) \]

This equation can predict the price of any house size within a similar range.


```python

w = model.coef_[0]
b = model.intercept_

print(f"Predicted Price Equation: y = {w:.2f}x + {b:.2f}")
```


## Step 5: Make Predictions

Let's predict the price of a **1,300 sq. ft. house**.


```python

new_size = np.array([[1300]])

predicted_price = model.predict(new_size)

print(f"Predicted price for 1300 sq. ft.: ${predicted_price[0]:,.2f}")
```


## Step 6: Visualize the Regression Line

The blue points represent actual data.

The red line represents the model predictions.


```python

predictions = model.predict(X)

plt.figure(figsize=(8,5))

plt.scatter(X, y, s=100, label='Actual Data')
plt.plot(X, predictions, linewidth=2, label='Regression Line')

plt.title('Linear Regression: House Size vs Price')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)

plt.show()
```


## Step 7: Evaluate the Model

Two common metrics:

### Mean Squared Error (MSE)

Measures average prediction error.

Lower MSE indicates better performance.

### R² Score

Measures how well the model explains the variation in the data.

- **R² = 1** → Perfect fit
- **R² = 0** → No explanatory power


```python

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```


## Summary

Linear Regression works best when:

- The relationship between variables is approximately linear.
- The target variable is numerical.
- The dataset contains minimal outliers.
- Features have low multicollinearity.
