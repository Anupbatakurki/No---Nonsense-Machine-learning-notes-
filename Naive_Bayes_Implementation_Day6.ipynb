# Naive Bayes Implementation (Day 6)

A beginner-friendly notebook explaining Multinomial Naive Bayes for spam detection.

## Step 1: Import Libraries

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

## Step 2: Create Sample Dataset

```
emails=[
"Free money now claim prize",
"Winner winner free gift",
"Meeting tomorrow at 3pm",
"Project deadline Friday",
"Cheap viagra buy now free",
"Lunch at noon today",
"Congratulations you won free",
"Report submission due"
]
labels=[1,1,0,0,1,0,1,0]
print(emails)
```

## Step 3: Convert Text to Numerical Features

```
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(emails)
print(vectorizer.get_feature_names_out())
print("Shape:",X.shape)
```

## Step 4: Train the Naive Bayes Model

```
model=MultinomialNB()
model.fit(X,labels)
print("Training Complete")
```

## Step 5: Predict a New Email

```
new_email=["Free winner prize claim now"]
new_X=vectorizer.transform(new_email)
prediction=model.predict(new_X)
probability=model.predict_proba(new_X)

print("Prediction:","SPAM" if prediction[0]==1 else "NOT SPAM")
print("Spam Probability:",probability[0][1])
print("Safe Probability:",probability[0][0])
```

## Step 6: Test Multiple Emails

Modify the emails below and observe the predictions.


```
tests=[
"Project meeting tomorrow",
"Congratulations you won free bitcoin",
"Lunch with friends",
"Claim your free gift now"
]
Xt=vectorizer.transform(tests)
pred=model.predict(Xt)
prob=model.predict_proba(Xt)

for e,p,pr in zip(tests,pred,prob):
    print("="*50)
    print(e)
    print("Prediction:","SPAM" if p==1 else "NOT SPAM")
    print("Spam Probability:",round(pr[1],3))
```

# Explanation

### Workflow
1. Import libraries.
2. Create a dataset.
3. Convert text into numbers using CountVectorizer.
4. Train Multinomial Naive Bayes.
5. Predict unseen emails.
6. Display prediction probabilities.

## Advantages
- Fast
- Simple
- Excellent for text classification

## Limitations
- Assumes features are independent.
- Less effective when features are highly correlated.
